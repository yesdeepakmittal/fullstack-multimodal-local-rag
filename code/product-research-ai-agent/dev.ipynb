```python
import os
import json
import requests
from pprint import pprint
from dotenv import load_dotenv
from helper import get_serpapi_url, get_data_from_serpapi

load_dotenv()

api_key = os.getenv("SERPAPI_API_KEY")
if not api_key:
    raise ValueError("SERPAPI_API_KEY environment variable is not set.")
```

```python
with open(os.path.join(os.getcwd(),"results", 'patent_1.json'), 'r') as f:
    data = json.load(f)
```

```python
# response.raise_for_status()  # Raise an error for bad responses
serpapi_url = get_serpapi_url(data)
response_data = get_data_from_serpapi(serpapi_url)

# Save the response data to a file
output_file = os.path.join(os.getcwd(), "results", "serpapi_response.json")
with open(output_file, 'w') as f:
    json.dump(response_data, f, indent=4)
```

```python
for idx, citation in enumerate(response_data.get('patent_citations', {}).get('original', {})):
    print(citation.get('serpapi_link', None))
```
